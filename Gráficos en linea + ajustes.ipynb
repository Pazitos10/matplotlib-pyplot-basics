```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
values = [] # Arreglo de valores
for i in range(100):
    values.append(np.random.normal(15,3)) #Generamos 10 valores random con distribucion normal
plt.subplot(1,2,1)
plt.plot(values)
plt.ylabel('Label Y')
plt.xlabel('Label X')
plt.subplot(1,2,2)
plt.ylabel('Otro Label Y')
plt.xlabel('Otro Label X')
plt.hist(values,5)
# No hace falta plt.show() para mostrar grafico
```

```python
#Puede observarse que los graficos se superponen, para lograr mejores resultados se usan grids
#Los grids pueden hacerse de 2 formas, con subplot2grid o con gridspec
#Ejemplo subplot2grid:
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3) #3 filas 3 columnas, primer grafico abarca las 3 col
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2) #segundo grafico, abarca 2 columnas
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2) #tercer grafico, ocupa 2 renglones
ax4 = plt.subplot2grid((3,3), (2, 0)) # 4to grafico ocupa lo que puede en columnas
ax5 = plt.subplot2grid((3,3), (2, 1)) # 5to grafico ocupa lo que puede en columnas
```

```python
#Ejemplo gridspec
import matplotlib.gridspec as gridspec
#esto:
ax = plt.subplot2grid((2,2),(0, 0))
#es equivalente a esto:
gs = gridspec.GridSpec(2, 2)
ax = plt.subplot(gs[0, 0])
#ax2 = plt.subplot(gs[1,:-1])
#ax3 = plt.subplot(gs[1:, -1])
```

```python
#El ejemplo de varios graficos, ahora con gridspec
gs = gridspec.GridSpec(3, 3)
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1,:-1])
ax3 = plt.subplot(gs[1:, -1])
ax4 = plt.subplot(gs[-1,0])
ax5 = plt.subplot(gs[-1,-2])
```

```python
#Realizando ajustes con metodo update aunque es posible definir esos valores en el constructor del Gridspec
# Atributos = left, right, wspace, hspace
# left = cuantas filas ocupa todo el gridspec
# right = cuantas columnas ocupa todo el gridspec
# hspace = espacio vertical
# wspace = espacio horizontal |<-->|
gs1 = gridspec.GridSpec(3, 3)
gs1.update(left=2, right=4, hspace=1.0, wspace=0.3)
#Es similar a subplots_adjust, pero solo afecta a los subplots creados por dicho GridSpec.

ax1 = plt.subplot(gs1[:-1, :])
plt.xlabel('ax1 X')
plt.ylabel('ax1 Y')
ax2 = plt.subplot(gs1[-1, :-1])
plt.xlabel('ax2 X')
plt.ylabel('ax2 Y')
ax3 = plt.subplot(gs1[-1, -1])
plt.xlabel('ax3 X')
plt.ylabel('ax3 Y')
```

```python
import calendar
dias = [np.array(calendar.mdays)[0:i].sum() + 1 for i in np.arange(12)+1]  # Para generar el lugar del primer días de cada mes en un año
meses = calendar.month_name[1:13]  # Creamos una lista con los nombres de los meses
plt.plot(np.arange(1,366), np.random.rand(365), label = 'valores al azar')  # Creamos un plot con 365 valores
plt.xlim(-5,370)  # Los valores del eje y variarán entre -5 y 370
plt.ylim(0,1.2)  # Los valores del eje y variarán entre 0 y 1.2
plt.legend()  # Creamos la caja con la leyenda
plt.title(u'Ejemplo de título')  # Ponemos un título
plt.suptitle(u'Ejemplo de título superior')  # Ponemos un título superior
plt.minorticks_on()  # Pedimos que se vean subrrayas de división en los ejes
plt.xticks(dias, meses, size = 'small', color = 'b', rotation = 45)  # Colocamos las etiquetas, meses, en las posiciones, dias, con color azul y rotadas 45º
plt.xlabel(u't (días)')
plt.ylabel('Media diaria')
```
